```python
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```python
img = plt.imread("lines.png")
```

```python
plt.imshow(img)
```

```python
im = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(im, cmap='gray')
```

```python
fx = [[-1, -1, -1], [0, 0, 0], [1, 1, 1] ]
m, n = im.shape
fx_im = im.copy().astype('float32')
im = im.astype('float32')
```

```python
for i in range(1, m-1):
    for j in range(1, n-1):
        fx_im[i,j] = im[i-1, j-1]*fx[0][0] + im[i-1, j]*fx[0][1] + im[i-1, j+1]*fx[0][2]+ im[i,j-1]*fx[1][0] + im[i,j]*fx[1][1] + im[i,j+1]*fx[1][2] + im[i+1,j-1]*fx[2][0] + im[i+1,j]*fx[2][1] + im[i+1,j+1]*fx[2][2]
```

```python
plt.imshow(fx_im, cmap='gray')
```

```python
plt.imshow(im, cmap='gray')
```

```python
fy = [[-1, 0, 1],
 [-1, 0, 1],
  [-1, 0, 1] ]
m, n = im.shape
fy_im = im.copy().astype('float32')
im = im.astype('float32')
```

```python
for i in range(1, m-1):
    for j in range(1, n-1):
        fy_im[i,j] = im[i-1, j-1]*fy[0][0] + im[i-1, j]*fy[0][1] + im[i-1, j+1]*fy[0][2]+ im[i,j-1]*fy[1][0] + im[i,j]*fy[1][1] + im[i,j+1]*fy[1][2] + im[i+1,j-1]*fy[2][0] + im[i+1,j]*fy[2][1]  + im[i+1,j+1]*fy[2][2]
```

```python
plt.imshow(fy_im, cmap='gray')
```

```python
adddded = fx_im + fy_im
```

```python
plt.imshow(adddded, cmap='gray')
```
