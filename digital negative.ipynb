**CONTRAST STRETCHING**

```python
a = 10
b = 20
l = 0.5
m = 5
n = 0.2
v = l * a
w = v + m * (b - a)
contrast = gray_image.copy()

for i in range(height):
    for j in range(width):
        if 0 < gray_image[i,j] <= a:
            contrast[i,j] = l*contrast[i,j]
        elif a < gray_image[i,j] <= b :
            contrast[i,j] = (m*contrast[i,j]-a) + v
        elif b < gray_image[i,j] <= 255:
            contrast[i,j] = (n*contrast[i,j]+b) + w
plt.imshow(contrast, cmap = 'gray')
```
