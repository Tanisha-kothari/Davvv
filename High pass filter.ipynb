```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cv2
```

```python
image = cv2.imread('image2.webp')
grey_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grey_img, cmap='gray')
```

```python
mask = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])
```

```python
m, n = grey_img.shape
final_hpf_image = grey_img.copy()
```

```python
for i in range(1, m-1):
    for j in range(1, n-1):
        final_hpf_image[i,j] = grey_img[i-1, j-1]*mask[0,0] + grey_img[i-1, j]*mask[0,1] + grey_img[i-1, j+1]*mask[0,2]+ grey_img[i,j-1]*mask[1,0] + grey_img[i,j]*mask[1,1]
        + grey_img[i,j+1]*mask[1,2] + grey_img[i+1,j-1]*mask[2,0] + grey_img[i+1,j]*mask[2,1] + grey_img[i+1,j+1]*mask[2,2]
```

```python
plt.imshow(final_hpf_image, cmap='gray')
```

```python

grey_img_new = cv2.imread('image2.webp', cv2.IMREAD_GRAYSCALE)

HBF = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

m, n = grey_img_new.shape
final_hbf_image_manual = grey_img_new.copy()

for i in range(1, m-1):
    for j in range(1, n-1):
        final_hbf_image_manual[i,j] = grey_img_new[i-1, j-1]*HBF[0,0] + grey_img_new[i-1, j]*HBF[0,1] + grey_img_new[i-1, j+1]*HBF[0,2] + \
                                      grey_img_new[i,j-1]*HBF[1,0] + grey_img_new[i,j]*HBF[1,1] + grey_img_new[i,j+1]*HBF[1,2] + \
                                      grey_img_new[i+1,j-1]*HBF[2,0] + grey_img_new[i+1,j]*HBF[2,1] + grey_img_new[i+1,j+1]*HBF[2,2]

plt.imshow(final_hbf_image_manual, cmap='gray')
```
