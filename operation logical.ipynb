```python
import cv2
import matplotlib.pyplot as plt
```

<b> <h1>Reading the image </b>

```python
img = cv2.imread('deku.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

<b> <h1>Displaying the image </b>

```python
plt.imshow(img, cmap=None)
```

<b> <h1>Observe the image </b>


```python
img.shape
```

```python
print(f"Height : {img.shape[:1]} , Width : {img.shape[1:2]}, Channels : {img.shape[2:3]}")
```

```python
b,g,r = cv2.split(img)
```

```python
plt.imshow(b)
```

```python
plt.imshow(g)
```

```python
plt.imshow(r)
```

  <b> <h1>Grayscale the image </b>

```python
plt.imshow(img, cmap='gray')
```

```python
roi = img[0 : 80, 0 : 700]
```

```python
plt.imshow(roi)
```

```python
addedimg = cv2.add(r, b)
plt.imshow(addedimg)
```

```python
addedimg = cv2.subtract(r, r)
plt.imshow(addedimg)
```

```python
addedimg = cv2.multiply(r, b)
plt.imshow(addedimg)
```

```python
andimg = cv2.bitwise_and(r,b)
plt.imshow(andimg)
```

```python
orimg = cv2.bitwise_or(r,b)
plt.imshow(orimg)
```

```python
xorimg = cv2.bitwise_xor(r,b)
plt.imshow(xorimg)
```
