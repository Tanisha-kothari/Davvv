# **IMAGE INPUT**

```python
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```python
img = plt.imread("1.webp")
```

```python
plt.imshow(img)
```

```python
im = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(im, cmap='gray')
```

# **HORIZONTAL MASK**

```python
mask_horizontal = [[-1, -1, -1], [2, 2, 2], [-1, -1, -1] ]
m, n = im.shape
hor_im = im.copy().astype('float32')
im = im.astype('float32')
```

```python
for i in range(1, m-1):
    for j in range(1, n-1):
        hor_im[i,j] = im[i-1, j-1]*mask_horizontal[0][0] + im[i-1, j]*mask_horizontal[0][1] + im[i-1, j+1]*mask_horizontal[0][2]+ im[i,j-1]*mask_horizontal[1][0] + im[i,j]*mask_horizontal[1][1] + im[i,j+1]*mask_horizontal[1][2] + im[i+1,j-1]*mask_horizontal[2][0] + im[i+1,j]*mask_horizontal[2][1] + im[i+1,j+1]*mask_horizontal[2][2]
```

```python
plt.imshow(hor_im, cmap='gray')
```

```python
plt.imshow(im, cmap='gray')
```

# **VERTICAL MASK**

```python
mask_vertical = [[-1, 2, -1],
 [-1, 2, -1],
  [-1, 2, -1] ]
m, n = im.shape
ver_im = im.copy().astype('float32')
im = im.astype('float32')
```

```python
for i in range(1, m-1):
    for j in range(1, n-1):
        ver_im[i,j] = im[i-1, j-1]*mask_vertical[0][0] + im[i-1, j]*mask_vertical[0][1] + im[i-1, j+1]*mask_vertical[0][2]+ im[i,j-1]*mask_vertical[1][0] + im[i,j]*mask_vertical[1][1] + im[i,j+1]*mask_vertical[1][2] + im[i+1,j-1]*mask_vertical[2][0] + im[i+1,j]*mask_vertical[2][1]  + im[i+1,j+1]*mask_vertical[2][2]
```

```python
plt.imshow(ver_im, cmap='gray')
```
