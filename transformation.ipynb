```python
import cv2
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
img = cv2.imread('dog.jpg')
im = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(im, cmap='gray')
```

# **TRANSLATION (in x, y and both)**

```python
tx = 100

translation_matrix = np.float32([
    [1, 0, tx],
    [0, 1, 0],
])

height, width = im.shape

translated_image = cv2.warpAffine(im, translation_matrix, (width, height))

plt.imshow(translated_image, cmap='gray')
```

```python
ty = 50

translation_matrix = np.float32([
    [1, 0, 0],
    [0, 1, ty],
])

height, width = im.shape

translated_image = cv2.warpAffine(im, translation_matrix, (width, height))

plt.imshow(translated_image, cmap='gray')
```

```python
tx, ty = 100, 50

translation_matrix = np.float32([
    [1, 0, tx],
    [0, 1, ty],
])

height, width = im.shape

translated_image = cv2.warpAffine(im, translation_matrix, (width, height))

plt.imshow(translated_image, cmap='gray')
```

# **ROTATION (clockwise - around origin, around a point, anticlockwise - around origin, around a point)**

```python
center_origin = (0, 0)
angle = 45
rotation_matrix_origin = cv2.getRotationMatrix2D(center_origin, angle, 1.0)
rotated_image_origin = cv2.warpAffine(im, rotation_matrix_origin, (width, height))

plt.imshow(rotated_image_origin, cmap='gray')
```

```python
point_of_rotation = (100, 50)
angle = 45
rotation_matrix_point = cv2.getRotationMatrix2D(point_of_rotation, angle, 1.0)
rotated_image_point = cv2.warpAffine(im, rotation_matrix_point, (width, height))

plt.imshow(rotated_image_point, cmap='gray')
```

```python
center_origin = (0, 0)
angle = -45
rotation_matrix_origin = cv2.getRotationMatrix2D(center_origin, angle, 1.0)
rotated_image_origin = cv2.warpAffine(im, rotation_matrix_origin, (width, height))

plt.imshow(rotated_image_origin, cmap='gray')
```

```python
point_of_rotation = (100, 50)
angle = -45
rotation_matrix_point = cv2.getRotationMatrix2D(point_of_rotation, angle, 1.0)
rotated_image_point = cv2.warpAffine(im, rotation_matrix_point, (width, height))

plt.imshow(rotated_image_point, cmap='gray')
```

# **SHEARING (in x, y, both)**

```python
import math
```

# **REFLECTION(ABOUT x, y, both)**

```python
rows, cols = im.shape
M = np.float32([[1,  0, 0],
                [0, -1, rows],
                [0,  0, 1]])
reflected_img = cv2.warpPerspective(im, M,(int(cols), int(rows)))
plt.imshow(reflected_img, cmap='gray')
```

```python
rows, cols = im.shape
M = np.float32([[-1,  0, cols],
                [0, 1, 0],
                [0,  0, 1]])

reflected_img = cv2.warpPerspective(im, M,(int(cols), int(rows)))
plt.imshow(reflected_img, cmap='gray')
```

# **1st Combination**

```python
angle_degrees = 30
angle_radians = math.radians(angle_degrees)
tangent_value = math.tan(angle_radians)

rows, cols = im.shape
M = np.float32([[1,  0, 0],
                [0, -1, rows],
                [0,  0, 1]])

shear_matrix = np.float32([
    [1, 0, 0],
    [tangent_value, 1, 0],
])

height, width = im.shape

shear_image = cv2.warpAffine(im, shear_matrix, (width, height))
reflected_img = cv2.warpPerspective(shear_image, M,(int(cols), int(rows)))
plt.imshow(reflected_img, cmap='gray')
```
